## Background:
### A manager at the bank is disturbed with more and more customers leaving their credit card services. They would really appreciate if one could predict for them who is gonna get churned so they can proactively go to the customer to provide them better services and turn customers' decisions in the opposite direction.


### In this notebook, we will download, explore and prepare data for further application of various machine learning methods to them. The results of these models will be presented in an analytical table for comparison.
### It will be interesting to see how different data preprocessing methods affect the performance of certain models (KNN, Decision tree, Support Vector Classifier, Random Forest).

<a id="Feature-Description"></a>
## Feature Description 

CLIENTNUM: Client number. Unique identifier for the customer holding the account

Customer_Age: Demographic variable - Customer's Age in Years

Gender: Demographic variable - M=Male, F=Female

Dependent_count: Demographic variable - Number of dependents

Education_Level: Demographic variable - Educational Qualification of the account holder (example: high school, college graduate, etc.)

Marital_Status: Demographic variable - Married, Single, Divorced, Unknown

Income_Category: Demographic variable - Annual Income Category of the account holder (< $40K, $40K - 60K, $60K - $80K, $80K-$120K, > $120K, Unknown)

Card_Category: Product Variable - Type of Card (Blue, Silver, Gold, Platinum)

Months_on_book: Period of relationship with bank

Total_Relationship_Count: Total no. of products held by the customer

Months_Inactive_12_mon: No. of months inactive in the last 12 months

Contacts_Count_12_mon: No. of Contacts in the last 12 months

Credit_Limit: Credit Limit on the Credit Card

Total_Revolving_Bal: Total Revolving Balance on the Credit Card

Avg_Open_To_Buy: Open to Buy Credit Line (Average of last 12 months)

Total_Amt_Chng_Q4_Q1: Change in Transaction Amount (Q4 over Q1)

Total_Trans_Amt: Total Transaction Amount (Last 12 months)

Total_Trans_Ct: Total Transaction Count (Last 12 months)

Total_Ct_Chng_Q4_Q1: Change in Transaction Count (Q4 over Q1)

Avg_Utilization_Ratio: Average Card Utilization Ratio

Attrition_Flag - Target

<h1 style="background-color:orange;font-family:newtimeroman;font-size:150%;text-align:center;">Data Preprocessing</h1>

```python
c_data.Attrition_Flag = c_data.Attrition_Flag.replace({'Attrited Customer': 1,'Existing Customer': 0})
c_data.Gender = c_data.Gender.replace({'F': 1,'M': 0})
c_data = pd.concat([c_data, pd.get_dummies(c_data['Education_Level']).drop(columns = ['Unknown'])], axis = 1)
c_data = pd.concat([c_data, pd.get_dummies(c_data['Income_Category']).drop(columns = ['Unknown'])], axis = 1)
c_data = pd.concat([c_data, pd.get_dummies(c_data['Marital_Status']).drop(columns = ['Unknown'])], axis = 1)
c_data = pd.concat([c_data, pd.get_dummies(c_data['Card_Category']).drop(columns = ['Platinum'])], axis = 1)
c_data.drop(columns = ['Education_Level', 'Income_Category', 'Marital_Status', 'Card_Category'], inplace = True)
c_data.head()
```

```python
raw_data = c_data.copy()
```

<h1 style="background-color:orange;font-family:newtimeroman;font-size:200%;text-align:center;">KNN (raw data)</h1>

### Prepare data for modeling

```python
y = raw_data['Attrition_Flag']
x = raw_data.drop('Attrition_Flag', axis = 1)
```

```python
x_train_80, x_test_20, y_train_80, y_test_20 = train_test_split(x, y, test_size=0.2, random_state = 0,
                                                                 shuffle= True, stratify = y)
print(y_train_80.value_counts())
print(y_test_20.value_counts())
```

### Let's try to find the optimal number of neighbors for the model. To do this, let's iterate over the values ​​in a certain range. We will select the optimal value by evaluating several metrics.

```python
k_num = range(1,40)
acc = []
acc_f1 = []
recall = []
roc_auc = []

for k in k_num:
    KNN = KNeighborsClassifier(n_neighbors = k).fit(x_train_80, y_train_80)
    y_hat = KNN.predict(x_test_20)
    acc.append(metrics.accuracy_score(y_test_20, y_hat))
    acc_f1.append(metrics.f1_score(y_test_20, y_hat, average='weighted'))
    recall.append(metrics.recall_score(y_test_20, y_hat))
    fpr1, tpr1, thresholds1 = metrics.roc_curve(y_test_20, y_hat)
    roc_auc.append(metrics.auc(fpr1, tpr1))
```

```python
fig, ax = plt.subplots(2, 2, figsize = (15, 10))
ax[0, 0].plot(range(1, 40), acc, 'b')
ax[0, 1].plot(range(1, 40), acc_f1, 'g')
ax[1, 0].plot(range(1, 40), recall, 'r')
ax[1, 1].plot(range(1, 40), roc_auc, 'r--')
ax[0, 0].set_ylabel('Accuracy')
ax[0, 0].set_xlabel('K')
ax[0, 1].set_ylabel('F1-score')
ax[0, 1].set_xlabel('K')
ax[1, 0].set_ylabel('Recall')
ax[1, 0].set_xlabel('K')
ax[1, 1].set_ylabel('AUC')
ax[1, 1].set_xlabel('K')
plt.show()
```

### By the graphs, we will make a choice in favor of 9 neighbors. Let us estimate the quality of the model using the confusion matrix.

```python
KNN = KNeighborsClassifier(n_neighbors = 9).fit(x_train_80, y_train_80)
y_hat = KNN.predict(x_test_20)

conf_mat = metrics.confusion_matrix(y_test_20, y_hat)
conf_mat = pd.DataFrame(conf_mat, index = KNN.classes_, columns = KNN.classes_)
conf_mat
```

### Let's try to apply the probability threshold approach - 'predict_proba'. We wil use only the metrics F-score and confusion matrix (due to the fact that the classes are unbalanced)

```python
k_num2 = range(1,40)
threshold_list = np.arange(0.05, 1, 0.05)
best = {}
d = {}
f1 = []

for k in k_num2:
    KNN2 = KNeighborsClassifier(n_neighbors = k).fit(x_train_80, y_train_80)
    pred_proba_df = pd.DataFrame(KNN2.predict_proba(x_test_20))
    for i in threshold_list:
        Y_test_pred = pred_proba_df.applymap(lambda x: 1 if x>i else 0)
        f1_score = metrics.f1_score(y_test_20, Y_test_pred.iloc[:,1], pos_label = 1, average='binary')        
        cm = metrics.confusion_matrix(y_test_20, Y_test_pred.iloc[:,1])
        d[f1_score] = cm
    best[k] = [max(d.keys()), d[max(d.keys())]]
    f1.append(max(d.keys()))
    d.clear()
```

```python
plt.figure()
g = sns.lineplot(x = k_num2, y = f1)
g.set_xlabel('K')
g.set_ylabel('f-score')
plt.show()
```

### Let's make a choice: K= 11

```python
KNN3 = KNeighborsClassifier(n_neighbors = 11).fit(x_train_80, y_train_80)
y_hat3 = KNN3.predict_proba(x_test_20)[:, 1]
#y_hat = KNN.predict(x_test_20)

fpr, tpr, thresholds = metrics.roc_curve(y_test_20, y_hat3)
roc_auc = metrics.auc(fpr, tpr)
```

```python
plt.figure(figsize = (12, 8))
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.4f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

### Let's try to define the optimal threshold

```python
pred_proba_df = pd.DataFrame(KNN3.predict_proba(x_test_20))
threshold_list = np.arange(0.025, 1, 0.025)
for i in threshold_list:
    print ('\n******** For i = {} ******'.format(i))
    Y_test_pred = pred_proba_df.applymap(lambda x: 1 if x>i else 0)
    #test_accuracy = metrics.accuracy_score(y_test_20, Y_test_pred.iloc[:,1])
    #print('Our testing accuracy is {}'.format(test_accuracy))
    f1_score = metrics.f1_score(y_test_20, Y_test_pred.iloc[:,1], pos_label=1, average='binary')
    print('Our testing f1_score is {}'.format(f1_score))

    print(metrics.confusion_matrix(y_test_20, Y_test_pred.iloc[:,1]))
    print(metrics.classification_report(y_test_20, Y_test_pred.iloc[:,1]))
```

```python
pt = preprocessing.PowerTransformer()
normalize_data_x = pt.fit_transform(X)

normalize_data_x = pd.DataFrame(normalize_data_x, columns = X.columns)
normalize_data_x.head() 
```
